```python
from pyecharts import Line

"""
attr 为横坐标
v1, v2 为两条不同线的纵坐标

add 方法的重要参数：

is_smooth 用于绘制平滑曲线
is_step 用于绘制阶梯折线图
is_fill 用于填充曲线下的面积
*_opacity 代表 * 相关的透明度, 1 为不透明, 0 为完全透明
    eg: area_opacity, line_opacity
symbol: 绘制图形时图标的表示 有'rect', 'roundRect', 'triangle', 'diamond', 'pin', 'arrow'可选
symbol_size 在柱状图中无效, 散点图中有效
"""
# Data generation
attr = ["衬衫", "羊毛衫", "雪纺衫", "裤子", "高跟鞋", "袜子"]
v1 = [5, 20, 36, 10, 10, 100]
v2 = [55, 60, 16, 20, 15, 80]
# 平滑曲线图
line = Line("折线图示例")
line.add("商家A", attr, v1, mark_point=["average"])
line.add("商家B", attr, v2, is_smooth=True, mark_line=["max", "average", ""])
line

# 阶梯图
line = Line("折线图-阶梯图示例")
line.add("商家A", attr, v1, is_step=True, is_label_show=True)
line

# 面积图
line = Line("折线图-面积图示例")
line.add("商家A", attr, v1, is_fill=True, line_opacity=0.2, area_opacity=0.5, symbol=None)
line.add("商家B", attr, v2, is_fill=True, area_color='#000', area_opacity=0.3, is_smooth=True)
line

# 堆叠图——可以显示全部的数据，也可将数据分开显示
# 包含了 symbol 的演示
line = Line("折线图-面积图示例")
line.add("商家A", attr, v1, is_stack=True, symbol='roundRect')
line.add("商家B", attr, v2, is_stack=True, area_color='#000', area_opacity=0.3, is_smooth=True)
line
```
