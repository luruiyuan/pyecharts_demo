```python
from pyecharts import Pie

# 使用面积展示数据大小
attr = ["衬衫", "羊毛衫", "雪纺衫", "裤子", "高跟鞋", "袜子"]
v1 = [11, 12, 60, 10, 10, 10]
pie_byArea = Pie("饼图示例1")
pie_byArea.add("", attr, v1, is_label_show=True, rosetype="area")
pie_byArea

# 使用圆心角展示数据大小
pie_radius = Pie("饼图示例2")
pie_radius.add("", attr, v1, is_label_show=True)
pie_radius
```
