```python
from pyecharts import Scatter

"""
散点图
"""

# v1 v2 分别为横纵坐标取值
v1 = [10, 20, 30, 40, 50, 60]
v2 = [10, 20, 30, 40, 50, 100]

# 简单使用
scatter = Scatter("散点图示例")
scatter.add("A", v1, v2)
scatter.add("B", v1[::-1], v2)
scatter

# 利用 Visualmap 来通过颜色映射数值
scatter = Scatter("散点图示例")
scatter.add("A", v1, v2)
scatter.add("B", v1[::-1], v2, is_visualmap=True)
scatter

# 利用 Visualmap 组件，通过图形点大小映射数值。
scatter = Scatter("散点图示例")
scatter.add("A", v1, v2)
scatter.add("B", v1[::-1], v2, is_visualmap=True, visual_type='size', visual_range_size=[10, 50])
scatter
```
