```python
import pandas as pd
import numpy as np
from pyecharts import Bar

# Data generation
index = pd.date_range('3/21/2017', periods=6, freq='M')
df1 = pd.DataFrame(np.random.randn(6), index=index)
df2 = pd.DataFrame(np.random.randn(6), index=index)

dtvalue1 = [i[0] for i in df1.values]
dtvalue2 = [i[0] for i in df2.values]
_index = [i for i in df1.index.format()]

# title 是主标题, subtitle 是副标题, 配色是自动的
# _index 是横坐标   dtvalue 是每一种 Bar 的纵坐标
# 用 add 方法可以添加, 每种柱子具体的设置用 add 方法的参数完成
# 例如 label_color 用于设置柱状图的柱子颜色, 其中 a->alpha 代表了透明度 0<=alpha<=1
title = 'bar chart'
subtitle = 'Profit and loss situation'
bar = Bar(title, subtitle)
bar.add('profit', _index, dtvalue1, label_color=['rgba(255,255,0,0.3)'])
bar.add('loss', _index, dtvalue2)
bar
```
